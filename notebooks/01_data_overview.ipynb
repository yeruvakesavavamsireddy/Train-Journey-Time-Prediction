```python
import pandas as pd

def load_data(path):
    df = pd.read_csv(path)
    return df

def dataset_summary(df):
    print("Shape:", df.shape)
    print("\nColumns:")
    print(df.columns)

    print("\nMissing Values:")
    print(df.isnull().sum())

    print("\nStatistics:")
    print(df.describe())

if __name__ == "__main__":
    df = load_data("data/raw/train_routes.csv")
    dataset_summary(df)
```
