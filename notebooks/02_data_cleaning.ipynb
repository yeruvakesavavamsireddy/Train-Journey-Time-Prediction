```python
import pandas as pd

def clean_data(df):

    df = df.drop_duplicates()

    df = df.dropna()

    df['arrival_time'] = pd.to_datetime(df['arrival_time'])
    df['departure_time'] = pd.to_datetime(df['departure_time'])

    df['journey_duration'] = (
        df['arrival_time'] - df['departure_time']
    ).dt.total_seconds() / 3600

    return df
```
