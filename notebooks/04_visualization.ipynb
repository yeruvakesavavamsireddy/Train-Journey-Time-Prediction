```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("data/raw/train_routes.csv")

sns.scatterplot(x="distance",y="journey_duration",data=df)

plt.savefig("reports/figures/distance_vs_time.png")

plt.clf()

df["station"].value_counts().plot(kind="bar")

plt.savefig("reports/figures/station_traffic.png")
```
